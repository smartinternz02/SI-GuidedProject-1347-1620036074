```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["name"]=le.fit_transform(data["name"])
```

```python
x=data.iloc[:,0:3].values
y=data.iloc[:,3:4].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
one=OneHotEncoder()
z=one.fit_transform(y[:,0:1]).toarray()
y=np.delete(y,0,axis=1)
y=np.concatenate((z,y),axis=1)
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
cat=Sequential()
```

```python
cat.add(Dense(3,kernel_initializer="random_uniform",activation="relu"))
```

```python
cat.add(Dense(6,kernel_initializer="random_uniform",activation="relu"))
```

```python
cat.add(Dense(8,kernel_initializer="random_uniform",activation="softmax"))
```

```python
cat.compile("adam","categorical_crossentropy",metrics=["accuracy"])
```

```python
cat.fit(x_train,y_train,batch_size=32,epochs=1500)
```

```python
ypred=cat.predict(x_test)
```
