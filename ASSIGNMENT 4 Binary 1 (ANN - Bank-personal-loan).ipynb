```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Bank_Personal_Loan_Modelling.csv")
```

```python
data.drop(columns=["ID","ZIP Code"],inplace=True)
```

```python
data.head()
```

```python
x=data.iloc[:,[0,1,2,3,4,5,6,8,9,10,11]].values
y=data.iloc[:,7].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=Sequential()
```

```python
ann.add(Dense(11,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.add(Dense(22,kernel_initializer="random_uniform",activation="relu"))
```

```python
ann.add(Dense(1,kernel_initializer="random_uniform",activation="sigmoid"))
```

```python
ann.compile("adam","binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=50)
```

```python
ypred=ann.predict(x_test)
```
