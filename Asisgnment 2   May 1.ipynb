```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,color="orange")
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=iris)
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,color="r")
```

```python
sns.lineplot(x="petal_length",y="petal_width",data=iris,color="g")
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```

```python
sns.displot(iris["species"])
```
